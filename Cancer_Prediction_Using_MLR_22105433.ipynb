# Reading CSV Cancer Data File

```r
cancer_data<-read.csv('cancer.csv')
cancer_data
```

### To verify if there is any null value present in the dataset 

```r
is.null(cancer_data)
```

### Deviding Data into training data and testing data

```r
set.seed(1)
sample_data <- sample(c(TRUE, FALSE), nrow(cancer_data), replace=TRUE, prob=c(0.7,0.3))
train_cancer_data  <- cancer_data[sample_data,]
test_cancer_data   <- cancer_data[!sample_data,]
```

### To check testing data summary(minimum value, maximum value, and quartiles of each column)

```r
summary(train_cancer_data)
```

## Building Machine learning model

### 1st Machine learning model

```r
model_1_cancer=lm(Death_Rate~.,data=cancer_data_int)
summary(model_1_cancer)
```

### Only including independent variables with high significance(3 asterix)

```r
model_1_cancer=lm(Death_Rate~Incidence_Rate+Average_Household_Size+Percentage_High_School_Age_18_24+Percentage_Bachelor_Degree_Age_25_Over+Percentage_Private_Coverage+Percentage_Employed_Private_Coverage+Percentage_Other_Race,data=cancer_data_int)
summary(model_1_cancer)
```
